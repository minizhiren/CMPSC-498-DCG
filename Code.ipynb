```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

datadir = "/content/drive/My Drive/Colab Notebooks/CMPSC498/HW1/data"
```

```python
# read data
df = pd.read_csv(f"{datadir}/YouTubeDataset_withChannelElapsed.csv")

# show header
print("Columns in the dataset:")
print(df.columns)
```

```python
# 计算数据集中有多少个唯一视频
num_videos = df["videoId"].nunique()

# 计算每个视频的总点赞数
likes_per_video = df.groupby("videoId")["videoLikeCount"].sum()

# 计算所有视频的平均点赞数
average_likes_per_video = likes_per_video.mean()

# 输出结果，四舍五入到 4 位小数
print(f"Average number of likes per video: {round(average_likes_per_video, 4)}")
```

```python
# 计算每个视频的互动率（Engagement Rate）
df["engagement_rate"] = (df["videoLikeCount"] + df["videoDislikeCount"] + df["VideoCommentCount"]) / df["videoViewCount"]

# 计算所有视频的平均互动率
average_engagement_rate = df["engagement_rate"].mean()

# 输出结果，四舍五入到 4 位小数
print(f"Average Engagement Rate across all videos: {round(average_engagement_rate, 4)}")
```

```python
# 已知：单个视频的元数据大小 = 500 字节
#       单个视频的缩略图大小 = 200 KB

# 单位转换：
metadata_size_MB = 500 / (1024 * 1024)  # 转换为 MB
thumbnail_size_MB = 200 / 1024  # 转换为 MB

# 计算单个视频的总存储大小
combined_video_size_MB = metadata_size_MB + thumbnail_size_MB

# 输出结果，四舍五入到 4 位小数
print(f"Combined storage size for a single video: {round(combined_video_size_MB, 4)} MB")
```

```python
# 计算总互动次数（engagements）
total_engagements = df["videoLikeCount"].sum() + df["videoDislikeCount"].sum() + df["VideoCommentCount"].sum()

# 计算所有互动产生的数据传输量
total_data_transfer_MB = total_engagements * combined_video_size_MB

# 输出结果，四舍五入到 4 位小数
print(f"Total data transfer for all engagements: {round(total_data_transfer_MB, 4)} MB")
```

```python
# 设数据集覆盖时间 = 5 年
years = 5
minutes_per_year = 365 * 24 * 60  # 一年中的总分钟数
total_minutes = years * minutes_per_year  # 5 年中的总分钟数

# 计算每分钟的观看次数
total_views = df["videoViewCount"].sum()
views_per_minute = total_views / total_minutes

# 输出结果，四舍五入到 4 位小数
print(f"Estimated number of views per minute: {round(views_per_minute, 4)}")
```

```python
# 计算每分钟访问数据的总量
data_accessed_per_minute_MB = views_per_minute * combined_video_size_MB

# 输出结果，四舍五入到 4 位小数
print(f"Estimated data accessed per minute: {round(data_accessed_per_minute_MB, 4)} MB")
```
